## style

1. label:
        plt.xlabel('')
        plt.ylabel('')
2. title:
        plt.title('')
3. thicks:
        plt.yticks([數值清單], 
                   [顯示名稱清單])

4. scale:
        plt.xscale('log')

### label, tittle, thick

### sizes, color, 透明度, 圖中文字, 格線
1. plt.scatter argument:
        plt.scatter(x =, y =, s =, c='red', alpha = 0.5)
        argument s, for size.  
        argument c, for color.
        argument alpha, for 透明度.
2. 圖中加入文字:
        plt.text(x座標, y座標, '文字')
3. grid:
        plt.grid(True)

```python
# Import numpy as np
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np.array(np_pop)**2.4

# Update: set s argument to np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)
#  c and alpha inside plt.scatter()
plt.scatter(x = gdp_cap, y = life_exp, s = np_pop**0.8, c= 'red', alpha = 0.5)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

# Add grid() call
plt.grid(True) 
# Display the plot
plt.show()
```
