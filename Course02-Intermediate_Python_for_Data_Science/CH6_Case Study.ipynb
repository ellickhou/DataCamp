## Random Number

* np.random.rand()
    * Pseudo-random Numbers
    * Mathematical formula
    * Starting from a seed
* np.random.seed()
    * Same seed → Same random number
    * Ensures ***Reproducibility***
* np.random.randint(0,2)
    * 0 to 1
    

```python
# Import numpy as np
import numpy as np

# Set the seed
np.random.seed(123)

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <=5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice, step)
```

## Random Walk

* 重複執行Random Number
* The current output will base on the previous one

```python
# Import numpy as np
import numpy as np

# Set the seed
np.random.seed(123)

# Initialize random_walk
random_walk = [0]

# for loop
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = max(0 , step - 1) # make step bever below 0
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='onedork')
%matplotlib inline

plt.clf()
# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

## Distribution

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(200) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        
        if np.random.rand() <= 0.001 :
            step = 0    
            
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

'''# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
plt.clf()'''

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()

#計算總步數，並以長條圖表示分布
# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()

#計算機率

chance_over_60 = np.count_nonzero(ends >= 60)/200
print(chance_over_60)
```
