## Adding a third variable with hue

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
file = 'student-alcohol-consumption.csv'
student_data= pd.read_csv(file)
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3",
                data=student_data, 
                hue="location", hue_order=['Rural', 'Urban'])

# Show plot
plt.show()
```

### violinplots and lvplots

```python
df = pd.read_csv('schoolimprovement2010grants.csv')
df=df.dropna()
# from matplotlib.axes._axes import _log as matplotlib_axes_logger
# matplotlib_axes_logger.setLevel('ERROR')
```

```python
fig ,(ax1,ax2) =plt.subplots(2,1, figsize=(6,12))
# Create a violinplot with the husl palette
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl',
              ax=ax1)
# Create a lvplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region',
              ax=ax2)
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,8))
# The final plot should include a line using a 2nd order polynomial
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",ax=ax[0,0])
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=5, ax=ax[1,0])
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            x_bins=8,
            order=2, ax=ax[1,1])
sns.regplot(data=df,
            y='Tuition',
            x="PCTPELL",
            fit_reg=False, ax=ax[0,1])

plt.show()
plt.clf()
```

## JointGrid, jointplot()
- `kind{ “scatter” | “reg” | “resid” | “kde” | “hex” }, optional`

```python
df = pd.read_csv('bike_share.csv')
```

```python
# Build a JointGrid comparing humidity and total_rentals

g = sns.JointGrid(x="hum",
            y="total_rentals",
            data=df,
            xlim=(0.1, 1.0)) 

g.plot(sns.regplot, sns.distplot)

plt.show()
plt.clf()
```

```python
# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x="temp",
         y="total_rentals",
         kind='reg',
         data=df,
         order=2,
         xlim=(0, 1))

plt.show()
plt.clf()
```

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="temp",
             y="casual",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
plt.clf()
```

```python
# Replicate the above plot but only for registered riders
g = (sns.jointplot(x="temp",
             y="registered",
             kind='hex',
             data=df,
             marginal_kws=dict(bins=10, rug=True, kde=True))
    .plot_joint(sns.scatterplot))

plt.show()
plt.clf()
```

```python
g = (sns.jointplot(x="temp",
             y="casual",
             kind='kde',
             data=df))#.plot_joint(sns.kdeplot))
    
plt.show()
plt.clf()
```

## Adding titles and labels
- seaborn object
    - FacetGrid
    - AxesSubplot
- title
    - FacetGrid.fig.suptitle
        - 大標題
    - FacetGrid.set_titles({col_name})
        - 小標題，col_name:seaborn 定義
    - AxesSubplot.set_title
        - 小標題
- label
    - .set(xlabel='', ylabel='')
    - plt.xticks(rotation=90)
        - Rotate x-tick labels

```python
sns.set_context("paper")
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                col='origin',
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower", y=1.03)
g.set_titles("{col_name}")
# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel='Car Model Year', ylabel='Average MPG')


# Show plot
plt.show()
```

## Customizing with matplotlib

```python
df=pd.read_csv('FY18_4050_FMRs.csv')
mean = df.fmr_1.describe()['mean']
median = df.fmr_1.describe()['50%']

# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=median, color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=mean, color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```
